#### **II - SUPPRIMONS LES COLONNES UNNAMED.**

```python
data.drop(['Unnamed: 5','Unnamed: 6',"Unnamed: 7","Unnamed: 8",'Unnamed: 9','Unnamed: 10'], axis=1, inplace=True)
data
```

#### **III - REMPLAÇONS LES VALEURS DE LA COLONNE WON? PAR DES 1 ET 0.**

```python
#RENOMMONS LA
data.rename(columns={"Won?": "Won"}, inplace=True)
```

```python
remplacements = {'NO':0, "YES":1, 'nan':'NAN'}
data['Won'] = data['Won'].map(remplacements)
data
```

#### **IV - NETOYONS LA COLONNE YEAR.**
#### OBJECTIF
**- Creer une colonne contenant chacune des editions de la cérémonie.**

**- Créer une colonne contenant des annees de remise de prix et supprimer la colonne year.**

```python
Years = data['Year'].str[0:4]
edition = data['Year'].str[:-3].str.split("(", expand=True)[1].astype(int)
```

```python
data['Years'] = Years
data['Edition'] = edition
data['Years'] = data['Years'].astype(int)
data.drop(['Year'], axis=1, inplace=True)
data
```

#### **V - NETOYONS LA COLONNE Additional Info.**
#### OBJECTIF
**- Creer une colonne contenant les film nominés .**

**- Créer une colonne contenant les personnages principaux de film et supprimer la colonne Additional Info.**

```python
split = data['Additional Info'].str.replace("'}", '').str.split("{'",expand=True)
split
```

```python
data['Movie'] = split[0]
data
```

```python
data['personnage'] = split[1]
data.drop(['Additional Info'], axis=1, inplace=True)
data
```

#### **VI - IDENTIFIONS ET SUPPRIMONS LES DOUBLONS.**

```python
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

#### **VII - ANALYSE DES VALEURS MANQUANTES.**

```python
#l'objectif est d'afficher le pourcentage par colonne de valeurs manquantes
(data.isna().sum()/len(data)*100).sort_values(ascending = False)
```

#### **VIII - EXPORTONS LE FICHIER EN CSV et SQL.**

```python
#EFFECTUONS UN TRI SUR LES DIFFERENTES EDITIONS
Clean_academy_award = data[['Edition','Years', 'Movie', 'personnage','Nominee','Category','Won']].sort_values("Edition")
Clean_academy_award
```

```python
Clean_academy_award.to_csv("ressources/Clean_academy_award.csv", index=False)
```

```python
con = sqlite3.connect("ressources/Clean_academy_award.db")
Clean_academy_award.to_sql("Clean_academy_award",con , index=False, if_exists = "replace")
```
